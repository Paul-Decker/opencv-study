### 图像阈值：cv2.threshold()

cv2.threshold() 函数是OpenCV库中用于图像阈值处理的一个函数。</br>阈值处理是一种基本的图像分割技术，其目的是将灰度图像转换为二值图像，或者将图像中的像素点根据某个阈值划分为不同的类。

##### ret, thresh = cv2.threshold(src, thresh, maxval, type)

参数:</br>
- src：表示输入图像（只能输入单通道图像），通常来说是灰度图像。
- thresh：阈值的具体值。
- maxval：当像素值超过了阈值thresh时，应该被赋予的新值。
- type：阈值类型，决定了如何处理图像中的像素值。
- - cv2.THRESH_BINARY：如果像素值大于阈值则设置为maxval，否则设置为0。
- - cv2.THRESH_BINARY_INV：如果像素值大于阈值则设置为0，否则设置为maxval。
- - cv2.THRESH_TRUNC：如果像素值大于阈值则设置为阈值，否则不改变。
- - cv2.THRESH_TOZERO：如果像素值大于阈值则不改变，否则设置为0。
- - cv2.THRESH_TOZERO_INV：如果像素值大于阈值则设置为0，否则不改变。
- - cv2.THRESH_OTSU：使用Otsu的二值化方法，自动计算阈值。
- - cv2.THRESH_TRIANGLE：使用三角形算法自动计算阈值。

返回值:</br>
- ret：返回的阈值。在某些情况下，这可能与输入的thresh相同，但在使用某些类型的阈值操作时，这个值会被调整为最优阈值。
- thresh：阈值操作后的输出图像。

##### 示例

```python
img = cv2.imread('assets/cat.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
ret, thresh1 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TRUNC) # truncation 截断
ret, thresh4 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO_INV)

titles = ['Original Image', 'BINARY', 'BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

# for i in range(6):
#     plt.subplot(2, 3, i + 1), plt.imshow(images[i], 'gray')
#     plt.title(titles[i])
#     plt.xticks([]), plt.yticks([])
# plt.show()

# 设置整个图形窗口的大小（单位是英寸）
fig = plt.figure(figsize=(12, 8))  # 假设想要一个宽度为10英寸，高度为8英寸的窗口
# 调整子图之间的间距
plt.subplots_adjust(left=0.1, right=0.9, top=0.7, bottom=0.1, hspace=0.5, wspace=0.5)
for i in range(6):
    img_rgb = cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB)
    plt.subplot(2, 3, i + 1)
    plt.imshow(img_rgb)
    plt.title(titles[i])
    # 清除坐标轴刻度
    plt.xticks([]) 
    plt.yticks([]) 
plt.show()
```
