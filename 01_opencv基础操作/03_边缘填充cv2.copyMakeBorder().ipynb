### 边缘填充

dst = cv2.copyMakeBorder(src, top, bottom, left, right, borderType, value)</br>
用于给图像添加边框的函数

参数:</br>
- src：输入图像，它应该是一个 numpy 数组。
- top：图像顶部边框的宽度。
- bottom：图像底部边框的宽度。
- left：图像左侧边框的宽度。
- right：图像右侧边框的宽度。
- borderType：定义边框类型的方法。以下是一些常用的边框类型：
- - cv2.BORDER_CONSTANT：添加恒定的彩色边框，颜色由 value 参数指定。
- - cv2.BORDER_REPLICATE：复制最边缘的像素填充边框。
- - cv2.BORDER_REFLECT：边框将是图像边界外的像素反射。（fedcba | abcdefgh | hgfedcb）
- - cv2.BORDER_REFLECT_101 或 cv2.BORDER_DEFAULT：边框将是图像边界外的像素反射，但不包括最边缘的像素。（gfedcb | abcdefgh | gfedcba）
- - cv2.BORDER_WRAP：边框像素将是对图像另一侧像素的包裹。（cdefgh | abcdefgh | abcdefg ）
- value：当 borderType 为 cv2.BORDER_CONSTANT 时，这个参数指定边框的颜色，它可以是单个数值（对于单通道图像）或者一个元组（对于多通道图像，如 (255, 0, 0) 表示红色）。

返回值:</br>
返回一个添加了边框的图像矩阵（ndarray数组）

```python
t, b, l, r = (50, 50, 50, 50)

replicate = cv2.copyMakeBorder(img, t, b, l, r, borderType=cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img, t, b, l, r, borderType=cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img, t, b, l, r, borderType=cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img, t, b, l, r, borderType=cv2.BORDER_WRAP)
constant = cv2.copyMakeBorder(img, t, b, l, r, borderType=cv2.BORDER_CONSTANT, value=0)
```

```python
# 2：表示子图分为2行  
# 3：表示子图分为3列  
# 1：表示当前操作的子图是整个子图布局中的第1个
plt.subplot(231), plt.imshow(img, 'gray'), plt.title('ORIGINAL')
plt.subplot(232), plt.imshow(replicate, 'gray'), plt.title('REPLICATE')
plt.subplot(233), plt.imshow(reflect, 'gray'), plt.title('REFLECT')
plt.subplot(234), plt.imshow(reflect101, 'gray'), plt.title('REFLECT_101')
plt.subplot(235), plt.imshow(wrap, 'gray'), plt.title('WRAP')
plt.subplot(236), plt.imshow(constant, 'gray'), plt.title('CONSTANT')

plt.show()
```
